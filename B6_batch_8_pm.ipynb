```python
# ----------------------------
# Grocery Store using OOP
# ----------------------------

class Product:
    def __init__(self, pid, name, price):
        self.pid = pid
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.pid}. {self.name} - {self.price}"


class Store:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def show_products(self):
        print("\nAvailable Products:")
        for product in self.products:
            print(product)

    def get_product_by_id(self, pid):
        for product in self.products:
            if product.pid == pid:
                return product
        return None


class Cart:
    def __init__(self):
        self.items = []

    def add_to_cart(self, product, qty):
        self.items.append((product, qty))

    def show_cart(self):
        print("\nYour Cart:")
        total = 0
        for product, qty in self.items:
            cost = product.price * qty
            print(f"{product.name} x {qty} = {cost}")
            total += cost
        print(f"Total: {total}")


# ----------------------------
# Main Program
# ----------------------------
store = Store()
store.add_product(Product(1, "Apples", 100))
store.add_product(Product(2, "Bananas", 50))
store.add_product(Product(3, "Milk", 60))
store.add_product(Product(4, "Bread", 40))

cart = Cart()

while True:
    store.show_products()
    try:
        choice = int(input("\nEnter Product ID to add to cart (0 to checkout): "))
        if choice == 0:
            break

        product = store.get_product_by_id(choice)
        if not product:
            print(" Invalid Product ID. Try again.")
            continue

        qty = int(input(f"Enter quantity for {product.name}: "))
        cart.add_to_cart(product, qty)
        print(f" {qty} {product.name} added to cart.")

    except ValueError:
        print(" Please enter a valid number.")

# Checkout
cart.show_cart()
print("\n Thank you for shopping with us!")
```
